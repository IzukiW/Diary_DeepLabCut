```python
#multicolumnsの作成
arrays=[
    ['x','x','x', 'y','y','y', 'diff','diff'],
    ['evaluation','mylabel','delta_x', 'evaluation','mylabel','delta_y', 'deviation', 'likelihood_eval']
]

tuples=list(zip(*arrays))
cols=pd.MultiIndex.from_tuples(tuples, names=['coords', 'label'])
# idx=df_eval.index
```

#### 編集ポイント4 (bodyparts 選択)

```python
bodyparts_list=['manubrium_base', 'mouth',	'prey',	'tentacle_base_1', 'tentacle_base_2', 'tentacle_base_3', 'tentacle_base_4', 'tentacle_base_5', 'tentacle_base_6', 'tentacle_base_7', 'tentacle_base_8','tentacle_base_9', 'tentacle_base_10', 'tentacle_base_11', 'tentacle_base_12', 'tentacle_base_13', 'tentacle_base_14', 'tentacle_base_15']
bdparts_df_list=[]
for i in bodyparts_list:
  name="df_diff" + "_"+ i
  bdparts_df_list.append(name)
print(bdparts_df_list)
```

```python
# dataframe作成

df_diff_mouth= pd.DataFrame(index=df_eval.index, columns=cols) #empty

df_diff_mouth.loc[:, ('x', 'evaluation')] = df_eval24.loc[:, idx['mouth', 'x']]
df_diff_mouth.loc[:, ('x', 'mylabel')] = df_mylabel24.loc[:, idx['mouth', 'x']]
df_diff_mouth.loc[:, ('y', 'evaluation')] = df_eval24.loc[:, idx['mouth', 'y']]
df_diff_mouth.loc[:, ('y', 'mylabel')] = df_mylabel24.loc[:, idx['mouth', 'y']]
df_diff_mouth.loc[:, ('x', 'delta_x')] = df_diff_mouth.loc[:, ('x', 'evaluation')] - df_diff_mouth.loc[:, ('x', 'mylabel')]
df_diff_mouth.loc[:, ('y', 'delta_y')] = df_diff_mouth.loc[:, ('y', 'evaluation')] - df_diff_mouth.loc[:, ('y', 'mylabel')]

df_diff_mouth.loc[:, ('diff', 'deviation')] =(df_diff_mouth.loc[:, ('x', 'delta_x')]**2 + df_diff_mouth.loc[:, ('y', 'delta_y')]**2)**0.5 #originalとevaluationのずれ
df_diff_mouth.loc[:, ('diff', 'likelihood_eval')]=df_eval24.loc[:, idx['mouth', 'likelihood']] #likelihood列追加
```
