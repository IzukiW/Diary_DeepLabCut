h5pyモジュールを使用する場合

```python
import h5py
import numpy as np

file_path = 'Cladonema_starved_crop_tentacle20_Izuki90shuffle1.mat' # MATLAB v7.3で保存された.matファイルのパス

try:
    with h5py.File(file_path, 'r') as f:
        # ファイル内のキー（変数名）を表示
        print("ファイル内のキー:", list(f.keys()))

        # 特定の変数にアクセス
        # MATLABの構造体（struct）は、h5pyではグループとして扱われることがあります。
        # データの取得方法は、.matファイル内の構造によって異なります。
        # 例: 'data_variable'という名前のデータがある場合
        if 'data_variable' in f:
            data = f['data_variable'][:] # データセット全体をNumPy配列として読み込む
            print("読み込んだデータ:", data)
            print("データの型:", type(data))
            print("データの形状:", data.shape)
        else:
            print("指定された変数 'data_variable' はファイル内に見つかりません。")

        # もしデータがオブジェクト参照として保存されている場合
        # MATLABで複雑なデータ型（例えば、cell配列やstring配列）を保存すると、
        # HDF5では直接のデータではなくオブジェクト参照として保存されることがあります。
        # その場合、以下のように参照を解決してデータを取得する必要があります。
        # 例えば、'my_cell_array'という変数名の場合:
        # if 'my_cell_array' in f:
        #     ref = f['my_cell_array'][()] # オブジェクト参照を取得
        #     if isinstance(ref, h5py.h5r.RegionReference):
        #         actual_data = f[ref][()] # 参照を解決してデータにアクセス
        #         print("セル配列データ:", actual_data)

except FileNotFoundError:
    print(f"エラー: ファイル '{file_path}' が見つかりません。")
except Exception as e:
    print(f"ファイルの読み込み中にエラーが発生しました: {e}")
```
